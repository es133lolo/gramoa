## Features

```python
feature_path = 'Processed/' + sub_folder_path + '/ADFTD/Feature'
if not os.path.exists(feature_path):
    os.makedirs(feature_path)
```

```python
path1 = 'Processed/' + sub_folder_path + '/ADFTD-RS/Feature'
path2 = 'Processed/' + sub_folder_path + '/ADFTD-PS/Feature'

sub_id = 1
for file1, file2 in zip(os.listdir(path1), os.listdir(path2)):
    print("Subject ID: ", sub_id)
    sub_path1 = os.path.join(path1, file1)
    sub_path2 = os.path.join(path2, file2)
    # concatenate the two features
    feature1 = np.load(sub_path1)
    feature2 = np.load(sub_path2)
    print(feature1.shape)
    print(feature2.shape)
    feature = np.concatenate((feature1, feature2), axis=0)
    print("Saved concatenated feature shape: ", feature.shape)
    np.save(feature_path + '/feature_{:02d}.npy'.format(sub_id), feature)
    sub_id += 1
    print("-------------------------------------\n")
```

```python
# Test the saved npy file
# example

path = feature_path

total_length = 0
for file in os.listdir(path):
    sub_path = os.path.join(path, file)
    print(np.load(sub_path).shape)
    total_length += np.load(sub_path).shape[0]
print("\nTotal length:", total_length)
```
